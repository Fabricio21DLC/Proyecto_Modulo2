Si el número aleatorio es menor a .5 entonces el paso es hacia la izquierda, y si es maro o igual a .5 será hacia la derecha.

```python
pasosDerecha = 0 
pasosIzquierda = 0
pasos = 1000
x0= 0
posiciones = [x0]
for paso in range(pasos):
    decision = np.random.rand()
    if decision < .5:
        pasosIzquierda += 1
        x0 -= 1
    else:
        pasosDerecha +=1
        x0 += 1
        posiciones.append(x0)
```

# Actividad

```python
def borracho(pasos):
    pasosDerecha = 0 
    pasosIzquierda = 0
    x0= 0
    posiciones = [x0]
    for paso in range(pasos):
        decision = np.random.rand()
        if decision < .4:
            pasosIzquierda += 1
            x0 -= 1
        else:
            pasosDerecha +=1
            x0 += 1
            posiciones.append(x0)
    return posiciones
```

```python
for r in range (5):
    plt.plot(borracho(1000), lw= .4, label = "Borracho %s" % r)
    plt.xlabel("Pasos")
    plt.ylabel("Posicion")
plt.legend()
plt.show()
```
