## Cuando se estira el resorte

$m\ddot{x}+kx+B\dot{x}=mg$

Requerimos la soluci贸n de

$\dot{x} = y$

$\dot{y} =g -\frac{k}{m}x - \frac{B}{m}y$

```python
y0=[.6,.8]
k=3.0
m=1.0
B=.5
g=32
```

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint
```

```python
def armonico(y,t,k,m,B):
    x=y[0]
    y=y[1]
    return [y, g -k*x/m -B/m * y]
```

```python
tiempo= np.linspace(0,50,500)
```

```python
solucion = odeint(armonico, y0, tiempo, args=(k,m,B))
```

```python
posicion = solucion[:,0]
velocidad = solucion[:,1]
```

```python
plt.figure(figsize=(8,6))
plt.plot(tiempo,posicion, c="r", lw= 2, label = "Posici贸n")
plt.plot(tiempo,velocidad, c="b", lw=2, label = "Velocidad")
plt.legend()
plt.xlabel("Tiempo")
plt.show()
```
