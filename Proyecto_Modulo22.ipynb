```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint
```

## Condiciones iniciales

- Altura del puente(Hp)= 200 metros
- Altura del deportista(Hd)= 1.8 metros
- $\beta$=1 (Resistencia del viento)
- g= -9.81 $\frac{m}{s^2}$
- P= 70 N
- m= 7.13 kg
- y(0)= 200 metros 
- y(t)=200-100-1.8= 98.2 metros

```python
y0=[200,0]
m=70
B=1
g=-9.81
```

$m\ddot{y}+B\dot{y}=mg$

Requerimos la soluci√≥n de

$\dot{y} = z$

$\dot{z} =g - \frac{B}{m}z$

```python
def funcion1(z,t,m,B,g):
    y=z[0]
    z=z[1]
    return [z, g -B*z/m]
```

```python
tiempo= np.linspace(0,5,6)
```

```python
solucion = odeint(funcion1, y0, tiempo, args=(m,B,g))
```

```python
posicion1 = solucion[:,0]
velocidad1 = solucion[:,1]
```

```python
plt.figure(figsize=(8,6))
plt.plot(tiempo,posicion1, c="r", lw= 2, label = "Posici√≥n")
plt.plot(tiempo,velocidad1, c="b", lw=2, label = "Velocidad")
plt.legend()
plt.xlabel("Tiempo")
plt.show()
```

## Cuando se estira el resorte

$m\ddot{y}+ky+B\dot{y}=mg$

Requerimos la soluci√≥n de

$\dot{y} = z$

$\dot{z} =g -\frac{k}{m}y - \frac{B}{m}z$

```python
y0=[200,0]
k=3
```

```python
def funcion2(z,t,k,m,g,B):
    y=z[0]
    z=z[1]
    return [z, g -(k*y/m) -(B/m)*z]
```

```python
tiempo= np.linspace(0,50,)
```

```python
solucion = odeint(funcion2, y0, tiempo, args=(k,m,g,B))
```

```python
posicion2 = solucion[:,0]
velocidad2 = solucion[:,1] 
```

```python
plt.figure(figsize=(8,6))
plt.plot(tiempo,posicion2, c="r", lw= 2, label = "Posici√≥n")
plt.plot(tiempo,velocidad2, c="b", lw=2, label = "Velocidad")
plt.legend()
plt.xlabel("Tiempo")
plt.show()
```
