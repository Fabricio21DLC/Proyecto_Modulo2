## Objetivo general

- Aplicar los conceptos de ecuaciones diferenciales vistos en clase para resolver la simulación de un salto en bungee.


## Objetivos específicos 

- Determinar la distancia y la velocidad del salto.
- Tomar en cuenta que se resolverán dos tipos de movimientos para x<=0 y x>0.

## Modelo a representar

Primero tenemos esta ecuación diferencial de segundo grado que representa el modelo que queremos simular (Caída libre):

$m\ddot{y}+B\dot{y}=mg$    


Luego la partimos en dos ecuaciones de primer grado para tener la posibilidad de resolverlas:  

$\dot{y} = z$

$\dot{z} =g - \frac{B}{m}z$

## Parametros a evaluar

- Altura del puente(Hp)= 200 metros

- Altura del deportista(Hd)= 1.8 metros

- $\beta$=1 (Resistencia del viento)

- g= -9.81 $\frac{m}{s^2}$

- P= 70 N

- m= 7.13 kg

- y(0)= 200 metros

- y(t)=200-100-1.8= 98.2 metros

# Simulaciones y gráficas

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint
```

```python
y0=[200,0]
m=70
B=1
g=-9.81
```

$m\ddot{y}+B\dot{y}=mg$

Requerimos la solución de

$\dot{y} = z$

$\dot{z} =g - \frac{B}{m}z$

```python
def funcion1(z,t,m,B,g):
    y=z[0]
    z=z[1]
    return [z, g -B*z/m]
```

```python
tiempo= np.linspace(0,5,6)
```

```python
solucion = odeint(funcion1, y0, tiempo, args=(m,B,g))
```

```python
posicion1 = solucion[:,0]
velocidad1 = solucion[:,1]
```

```python
plt.figure(figsize=(8,6))
plt.plot(tiempo,posicion1, c="r", lw= 2, label = "Posición")
plt.plot(tiempo,velocidad1, c="b", lw=2, label = "Velocidad")
plt.legend()
plt.xlabel("Tiempo")
plt.show()
```

## Cuando se estira el resorte :

$m\ddot{y}+ky+B\dot{y}=mg$

Requerimos la solución de

$\dot{y} = z$

$\dot{z} =g -\frac{k}{m}y - \frac{B}{m}z$

```python
y0=[200,0]
k=14
```

```python
def funcion2(z,t,k,m,g,B):
    y=z[0]
    z=z[1]
    return [z, g -(k*y/m) -(B/m)*z]
```

```python
tiempo= np.linspace(0,50,51)
```

```python
solucion = odeint(funcion2, y0, tiempo, args=(k,m,g,B))
```

```python
posicion2 = solucion[:,0]
velocidad2 = solucion[:,1]
```

```python
plt.figure(figsize=(8,6))
plt.plot(tiempo,posicion2, c="r", lw= 2, label = "Posición")
plt.plot(tiempo,velocidad2, c="b", lw=2, label = "Velocidad")
plt.legend()
plt.xlabel("Tiempo")
plt.show()
```

## Espacio Fase 1:

```python
plt.figure(figsize=(5,5))
plt.plot(posicion1,velocidad1, lw= 2,);
plt.xlabel("Posición")
plt.ylabel("Velocidad")
plt.title("Espacio Fase")
plt.show()
```

## Espacio fase 2:

```python
plt.figure(figsize=(5,5))
plt.plot(posicion2,velocidad2, lw= 2,);
plt.xlabel("Posición")
plt.ylabel("Velocidad")
plt.title("Espacio Fase")
plt.show()
```
